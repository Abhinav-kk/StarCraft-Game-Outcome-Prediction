# Visualization of Processed Dataset

### Visualising PvP dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../1D_All_ReplaysData_PvP.csv')
```

```python
data.describe()
```

```python
data.head()
```

```python
# get last row of each replay
single_replays = data.groupby('ReplayID').last()
```

```python
num_replays = len(list(data.groupby('ReplayID')))
num_replays
```

```python
# Plotting Duration of Replays
bin_edges = plt.hist(single_replays['Duration'], bins=30)[1]  # Returns bin edges
plt.xlabel('Duration (s)')
plt.ylabel('Frequency')
plt.title('Duration of Replays')
plt.grid(axis='y')  # Add grid lines for y-axis
# Set x-ticks to bin edges - for clarity, you might want to only display a subset of these
plt.xticks(bin_edges, rotation=90)  
plt.show()
```

Most Matches in PvP game replays are of duration 13000 to 42000 Frames.
There are some outliers of games with higher duration than 100000 Frames.

```python
single_replays.sort_values(by='Duration', ascending=False)
```

### Top 20 Maps for PvP Matches

```python
# Plotting Bar graph of map names

map_counts = single_replays['MapName'].value_counts().sort_values(ascending=False)[:20]
# Plotting
plt.figure(figsize=(10, 6))  # Increase figure width for better spacing
map_counts.plot(kind='bar')
plt.title('Number of Games Played on Each Map')
plt.xlabel('Map Name')
plt.ylabel('Number of Games')
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
map_counts
```

# Pie chart of top 50 map names

```python
# Plotting pie chart graph of game maps
plt.figure(figsize=(18, 28))
map_counts = single_replays['MapName'].value_counts()
map_counts.sort_values(ascending=False).head(50).plot(kind='pie', autopct='%1.1f%%')
plt.title('Percentage of Games Played on Each Map (Top 50)')
plt.ylabel('')
plt.show()
```

### Bar graph of top 50 played maps

```python

plt.figure(figsize=(15, 10))
map_counts.sort_values(ascending=False).head(50).plot(kind='bar')
plt.title('Number of Games Played on Each Map (Top 50)')
plt.xlabel('Map Name')
plt.ylabel('Number of Games')
plt.tight_layout()
plt.show()
```

## Map Dimensions Visualization

```python
# Plotting bar graph map width and height
plt.figure(figsize=(10, 5))
single_replays['MapWidth'].value_counts().plot(kind='bar')
plt.title('Number of Maps with Each Width')
plt.xlabel('Map Width')
plt.ylabel('Number of Maps')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))
single_replays['MapHeight'].value_counts().plot(kind='bar')
plt.title('Number of Maps with Each Height')
plt.xlabel('Map Height')
plt.ylabel('Number of Maps')
plt.tight_layout()
plt.show()
```

```python
# Pie Chart for map width and height with subplot
fig, ax = plt.subplots(1, 2, figsize=(15, 7))
single_replays['MapWidth'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=ax[0])
ax[0].set_title('Percentage of Maps with Each Width')
ax[0].set_ylabel('')
single_replays['MapHeight'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=ax[1])
ax[1].set_title('Percentage of Maps with Each Height')
ax[1].set_ylabel('')
plt.show()
```

### Distribution of Actions Per Minute of Each Player

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate APM for each player
player1_apm = single_replays['Player1_EAPM']
player2_apm = single_replays['Player2_EAPM']

# Plotting
plt.figure(figsize=(10, 6))
sns.histplot(player1_apm, color='blue', kde=True, label='Player 1 APM', alpha=0.5)
sns.histplot(player2_apm, color='red', kde=True, label='Player 2 APM', alpha=0.5)
plt.title('Distribution of Actions Per Minute for Each Player')
plt.xlabel('Actions Per Minute (APM)')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

This shows that the both players number of actions per minute are close in density but player 1 has a higher peak than player 2.

### Distribution of Effective Commands of Each Player

```python
# Plotting ECMD Count for each player
player1_ecmd = single_replays['Player1_ECmdCount']
player2_ecmd = single_replays['Player2_ECmdCount']

# Plotting
plt.figure(figsize=(10, 6))
sns.histplot(player1_ecmd, color='blue', kde=True, label='Player 1 ECMD', alpha=0.5)
sns.histplot(player2_ecmd, color='red', kde=True, label='Player 2 ECMD', alpha=0.5)
plt.title('Distribution of ECMD Count for Each Player')
plt.xlabel('ECMD Count')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Remove 0 Winners
valid_replays = single_replays[single_replays['Winner'] != 0]
```

```python
# Plot Winner Distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=single_replays['Winner'].value_counts().index, y=single_replays['Winner'].value_counts())
plt.title('Distribution of Winners')
plt.xlabel('Winner')
plt.ylabel('Count')
plt.show()
```
